```python
train_data_path = "/kaggle/working/train/"
test_data_path = "/kaggle/working/test1"

train_images = os.listdir(train_data_path)
test_images = os.listdir(test_data_path)

#checking the dataset
print(train_images[:5])
print(test_images[:5])

dog_images = []
cat_images = []

for file_names in train_images:
    split_name = file_names.split(".")
    if split_name[0] == "dog": dog_images.append(file_names);
    else: cat_images.append(file_names);
```

```python
print(cat_images[:5])
print(dog_images[:5])

import cv2
import matplotlib.pyplot as plt

train_images = []
Labels = []

dir_name = "/kaggle/working/train/"
for cats in cat_images:
    image_read = cv2.imread(os.path.join(dir_name,cats));
    image_read = cv2.resize(image_read,(224,224))
    train_images.append(image_read)
    Labels.append(0)

print(len(train_images))

#reducing the dataset
train_images = train_images[:10000]
Labels = Labels[:10000]

for dogs in dog_images:
    image_read = cv2.imread(os.path.join(dir_name,dogs));
    image_read = cv2.resize(image_read,(224,224))
    train_images.append(image_read)
    Labels.append(1)

train_images = train_images[:20000]
Labels = Labels[:20000]

print(len(train_images))
```

```python
#for layer in model_vgg16.layers:
    #layer.trainable = False
    #print(layer, layer.trainable)
  
from tensorflow import keras
import tensorflow.keras.layers as Layers
from skimage.color import rgb2gray

model = keras.Sequential()
model.add(keras.layers.Conv2D(250,kernel_size=(3,3),activation='relu',input_shape=(224,224,3)))
model.add(keras.layers.MaxPool2D(5,5))
model.add(keras.layers.Conv2D(100,kernel_size=(3,3),activation='relu'))
model.add(keras.layers.MaxPool2D(5,5))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(1024, activation='relu'))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(3, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['acc'])

print(train_images.shape)
print(len(Labels))
training = model.fit(rgb2gray(train_images),Labels,epochs=35)
```
